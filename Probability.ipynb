### Probability and Statistics

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
```

```python
# Create a function to plot the probability density function
def plot_pdf(x, y, title, x_label, y_label, color):
    plt.plot(x, y, color=color)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()

# Create a function to plot the cumulative density function
def plot_cdf(x, y, title, x_label, y_label, color):
    plt.plot(x, y, color=color)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()

# Create a function to plot the histogram
def plot_hist(x, title, x_label, y_label, color):
    plt.hist(x, color=color)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
```

```python
# Create and plot a normal distribution
x = np.linspace(-5, 5, 100)
y = stats.norm.pdf(x, 0, 1)
plot_pdf(x, y, 'Normal Distribution', 'x', 'Probability Density', 'blue')
```

```python
# Create a function to plot multiple distributions for comparison
def plot_distributions(x, y1, y2, y3, title, x_label, y_label, color1, color2, color3):
    plt.plot(x, y1, color=color1)
    plt.plot(x, y2, color=color2)
    plt.plot(x, y3, color=color3)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()

# Create and plot multiple distributions for comparison
x = np.linspace(-5, 5, 100)
y1 = stats.norm.pdf(x, 0, 1)
y2 = stats.uniform.pdf(x, -2, 4)
y3 = stats.norm.pdf(x, 0, 2)
plot_distributions(x, y1, y2, y3, 'Normal, Uniform, and Normal Distributions', 'x', 'Probability Density', 'blue', 'red', 'green')
```

```python
# Create a function to calculate the expectation and variance of a random variable
def expectation_var(x):
    # Calculate the expectation
    expectation = np.mean(x)
    # Calculate the variance
    variance = np.var(x)
    return expectation, variance

# Create a function to calculate the probability density function
def pdf(x, mu, sigma):
    # Calculate the probability density function
    pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
    return pdf
```

```python
# Write the test for the pdf function and expectation_var function
def test_pdf():
    # Create a random variable
    x = np.random.normal(0, 1, 1000)
    # Calculate the expectation and variance
    expectation, variance = expectation_var(x)
    # Calculate the probability density function
    y = pdf(x, expectation, variance)
    # Plot the probability density function
    plot_pdf(x, y, 'Probability Density Function', 'x', 'pdf(x)', 'blue')

def test_expectation_var():
    # Create a random variable
    x = np.random.normal(0, 1, 1000)
    # Calculate the expectation and variance
    expectation, variance = expectation_var(x)
    print('The expectation is: ', expectation)
    print('The variance is: ', variance)

test_expectation_var()
test_pdf()
```

```python
# Create a function to calculate and display covariance and correlation
def covariance_correlation(x, y):
    # Calculate the covariance
    covariance = np.cov(x, y)
    # Calculate the correlation
    correlation = np.corrcoef(x, y)
    print('The covariance is: ', covariance)
    print('The correlation is: ', correlation)

# Write the test for the covariance_correlation function
def test_covariance_correlation():
    # Create a random variable
    x = np.random.normal(0, 1, 1000)
    y = np.random.normal(0, 1, 1000)
    # Calculate the covariance and correlation
    covariance_correlation(x, y)

test_covariance_correlation()
```
